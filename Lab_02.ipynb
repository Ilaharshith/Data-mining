```python
data = pd.DataFrame({
    'Math':[85, 78, 92, 88, 76],
    'Science':[88, 79, 95, 84, 82],
    'English':[75, 83, 89, 92, 80]
})

#correlation matrix
correlation_matrix = data.corr()

#heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```

```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic data (replace this with your dataset)
# Create a DataFrame
data = {
    'Time': ['2025-01-01', '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-05'],
    'Value': [10, 20, 15, 25, 30]
}
df = pd.DataFrame(data)

# If Time is in string format, convert it to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(df['Time'], df['Value'], marker='o', color='b', label='Value Over Time')

# Add titles and labels
plt.title('Line Plot Example')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
import pandas as pd

# Step 2: Prepare data
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [23, 45, 56, 78]}
df = pd.DataFrame(data)

# Step 3: Create bar plot
plt.figure(figsize=(8, 5))  # Set the figure size
plt.bar(df['Category'], df['Values'], color='skyblue')

# Step 4: Add labels and title
plt.xlabel('Category', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.title('Bar Plot Example', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 5: Display the plot
plt.show()
```

```python
# Step 1: Import Libraries
import matplotlib.pyplot as plt

# Step 2: Prepare Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [45, 30, 15, 10]

# Step 3: Create a Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'lightgreen', 'pink'])

# Step 4: Add Title
plt.title('Pie Chart Example', fontsize=16)

# Step 5: Display the Chart
plt.show()
```
